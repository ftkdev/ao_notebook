```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load parsed Australian Open Data
url = "https://raw.githubusercontent.com/ftkdev/ao_scraper_project/master/australian_open/extracts/aodata_final.csv"
df = pd.read_csv(url)
```

```python
# Fix calendar_year column
df['calendar_year'] = df['calendar_year'].str.replace('[', '', regex=True).str.replace(']', '', regex=True).astype(int)
df['calendar_year']
```

```python
# Check db integrity
df[['calendar_year', 'tournament_phase', 'tournament']].pivot_table(index='calendar_year', 
               columns='tournament_phase', aggfunc='count')
```

```python
# Top 20 players with most average per match aces in Australian Opens since 2016
df_p1 = df[['calendar_year', 'player1_name', 'player1_surname', 'p1_aces']]
df_p2 = df[['calendar_year', 'player2_name', 'player2_surname', 'p2_aces']]
df_p2.columns = df_p1.columns
df_aces = pd.concat([df_p1, df_p2], ignore_index=True)
df_aces = df_aces.rename(columns={'player1_name': 'player_name', 
                                  'player1_surname': 'player_surname', 'p1_aces': 'aces'})
df_aces = df_aces.dropna()

filt = (df_aces['calendar_year'] > 2015)
df_aces = df_aces.loc[filt , :]

df_aces.pivot_table(index=['player_surname', 'player_name'], columns = 'calendar_year',
                    values='aces', aggfunc=np.average, margins=True).sort_values(by='All', ascending=False).head(20)
```

```python
# Top 20 most aces in Australian Opens since 2016 chart
df_aces_chart = df_aces.pivot_table(index=['player_surname', 'player_name'],
                    values='aces', aggfunc=np.sum).sort_values(by='aces', ascending=False).head(20)

df_aces_chart.plot.bar(figsize=(14,7), title='AO Top 20 Aces', fontsize = 15)
```

```python
# Top 20 players on tiebreaks won since 2016
df['p1_tiebreakwin'] = (
    np.where(((df['player1_1st_set'] == 7) & (df['player2_1st_set'] == 6)), 1, 0) +
    np.where(((df['player1_2nd_set'] == 7) & (df['player2_2nd_set'] == 6)), 1, 0) +
    np.where(((df['player1_3rd_set'] == 7) & (df['player2_3rd_set'] == 6)), 1, 0) +
    np.where(((df['player1_4th_set'] == 7) & (df['player2_4th_set'] == 6)), 1, 0) +
    np.where(((df['player1_5th_set'] == 7) & (df['player2_5th_set'] == 6)), 1, 0)
)

df['p2_tiebreakwin'] = (
    np.where(((df['player1_1st_set']==6) & (df['player2_1st_set'] == 7)), 1, 0) +
    np.where(((df['player1_2nd_set']==6) & (df['player2_2nd_set'] == 7)), 1, 0) +
    np.where(((df['player1_3rd_set']==6) & (df['player2_3rd_set'] == 7)), 1, 0) +
    np.where(((df['player1_4th_set']==6) & (df['player2_4th_set'] == 7)), 1, 0) +
    np.where(((df['player1_5th_set']==6) & (df['player2_5th_set'] == 7)), 1, 0)
)

df['tiebreaks'] = (
    np.where((df['player1_1st_set']+df['player2_1st_set'] == 13), 1, 0) +
    np.where((df['player1_2nd_set']+df['player2_2nd_set'] == 13), 1, 0) +
    np.where((df['player1_3rd_set']+df['player2_3rd_set'] == 13), 1, 0) +
    np.where((df['player1_4th_set']+df['player2_4th_set'] == 13), 1, 0) +
    np.where((df['player1_5th_set']+df['player2_5th_set'] == 13), 1, 0)
)

df_p1 = df[['calendar_year', 'player1_name', 'player1_surname', 'p1_tiebreakwin', 'tiebreaks']]
df_p2 = df[['calendar_year', 'player2_name', 'player2_surname', 'p2_tiebreakwin', 'tiebreaks']]
df_p2.columns = df_p1.columns
df_tiebreakwin = pd.concat([df_p1, df_p2], ignore_index=True)
df_tiebreakwin = df_tiebreakwin.rename(columns={'player1_name': 'player_name', 
                                  'player1_surname': 'player_surname', 'p1_tiebreakwin': 'tiebreakwin'})
df_tiebreakwin = df_tiebreakwin.dropna()

filt = (df_tiebreakwin['calendar_year'] > 2015)
df_tiebreakwin = df_tiebreakwin.loc[filt , :]

df_tiebreakwin_pivot = df_tiebreakwin.pivot_table(index=['player_surname', 'player_name'],
                    values=['tiebreakwin', 'tiebreaks'], aggfunc=np.sum)

df_tiebreakwin_pivot['tiebreaks%'] = df_tiebreakwin_pivot['tiebreakwin'] / df_tiebreakwin_pivot['tiebreaks']

df_tiebreakwin_pivot.sort_values(by='tiebreakwin', ascending=False).head(20)
```
